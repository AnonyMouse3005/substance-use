```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re
import json
from IPython.display import display_html
import statistics as stats
```

```python
with open('mappings.json', 'r') as f:
    mappings = json.load(f)

# mappings = {re.sub('[^A-Za-z0-9]+$', '', k): v for k, v in mappings.items()}

with open('wave1_vars.json', 'r') as f:
    dict1 = json.load(f)

with open('wave2_vars.json', 'r') as f:
    dict2 = json.load(f)
```

```python
df_C1W1nonet = pd.read_csv('C1W1_nonnetwork.csv')
df_C1W2nonet = pd.read_csv('C1W2_nonnetwork.csv')
df_C1W1net = pd.read_csv('C1W1_network.csv')
```

```python
def describe_df(df):
    d = df.describe().loc[['count','mean','std','min','max']]
    return pd.concat([d, df.reindex(d.columns, axis = 1).agg([np.median, stats.mode])])
```

```python
for key in mappings.keys():
    val = mappings[key]
    if key[-1] != '^':  # exclude text variables
        df1 = describe_df(pd.DataFrame(df_C1W1nonet.iloc[:,val[2]]))
        df1_styler = df1.style.set_table_attributes("style='display:inline'")
        df2 = describe_df(pd.DataFrame(df_C1W2nonet.iloc[:,val[3]]))
        df2_styler = df2.style.set_table_attributes("style='display:inline'")
        print(val[1])
        display_html(df1_styler._repr_html_()+df2_styler._repr_html_(), raw=True)
        print('--------------------------------------------------------------------------')
```

```python
W1_exckeys = [col for col in df_C1W1nonet if col.startswith('AC')]
W1_exckeys.extend(['CJ4','CJ5','CJ6','CJ7'])
for key in W1_exckeys:
    df = pd.DataFrame(df_C1W1nonet[key])
    d = describe_df(df)
    print(dict1[key])
    display_html(d.style)
    print('--------------------------------------------------------------------------')
```

```python
W2_exckeys = ['Q70','Q199']
for key in W2_exckeys:
    df = pd.DataFrame(df_C1W2nonet[key])
    d = describe_df(df)
    print(dict2[key])
    display_html(d.style)
    print('--------------------------------------------------------------------------')
```
